Q1=> What is NumPy, and why is it widely used in Python?

Q2=> How does broadcasting work in NumPy?

Aans=>Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes without making explicit copies. It works by expanding the smaller array along dimensions where necessary to match the shape of the larger array.

```python


# Q4=> How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset('tips')


df['smoker'] = df['smoker'].map({'Yes': 1, 'No': 0})

corr_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")

plt.show()

```

```python

# Q5=> Generate a bar plot using Plotly?

import plotly.express as px
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [10, 15, 7, 12, 20]
}

df = pd.DataFrame(data)


fig = px.bar(df, x='Category', y='Values', title="Sample Bar Plot",
             labels={'Values': 'Count', 'Category': 'Categories'},
             color='Category')


fig.show()


```

```python

# Q8=> Create a line plot with multiple lines using Matplotlib?


import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


plt.figure(figsize=(8, 6))


plt.plot(x, y1, label='sin(x)', linestyle='-', color='b')
plt.plot(x, y2, label='cos(x)', linestyle='--', color='g')
plt.plot(x[np.abs(y3) < 5], y3[np.abs(y3) < 5], label='tan(x)', linestyle=':', color='r')


plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")

plt.grid(True)
plt.legend()


plt.show()

```

```python

# Q9=> Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?


import pandas as pd


data = {'A': [10, 20, 30, 40, 50],
        'B': [15, 25, 35, 45, 55]}
df = pd.DataFrame(data)


threshold = 30


filtered_df = df[df['A'] > threshold]

print(filtered_df)
```

```python

# Q13=> Create a 3D scatter plot using Plotly


import plotly.express as px
import pandas as pd
import numpy as np


n = 100
df = pd.DataFrame({
    'x': np.random.randn(n),
    'y': np.random.randn(n),
    'z': np.random.randn(n),
    'color': np.random.choice(['red', 'blue', 'green'], size=n)
})


fig = px.scatter_3d(df, x='x', y='y', z='z', color='color')

fig.show()
```
